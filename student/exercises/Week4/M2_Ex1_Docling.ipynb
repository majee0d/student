```python
from pathlib import Path
import json
import yaml
import time
import pandas as pd

from docling.document_converter import DocumentConverter
```

```python
DATA_DIR = Path("docling_inputs")
OUTPUT_DIR = Path("scratch")

DATA_DIR.mkdir(exist_ok=True)
OUTPUT_DIR.mkdir(exist_ok=True)

print("Input folder:", DATA_DIR.resolve())
print("Output folder:", OUTPUT_DIR.resolve())
```

```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from docx import Document

# PDF 1
pdf1 = DATA_DIR / "sample_ai.pdf"
c = canvas.Canvas(str(pdf1), pagesize=letter)
c.setFont("Helvetica", 14)
c.drawString(72, 720, "Artificial Intelligence Report")
c.setFont("Helvetica", 11)
c.drawString(72, 690, "AI helps machines perform tasks that usually require human intelligence.")
c.drawString(72, 670, "This PDF is used for Docling conversion practice.")
c.save()

# PDF 2
pdf2 = DATA_DIR / "sample_data.pdf"
c = canvas.Canvas(str(pdf2), pagesize=letter)
c.setFont("Helvetica", 14)
c.drawString(72, 720, "Data Science Notes")
c.setFont("Helvetica", 11)
c.drawString(72, 690, "Data science combines statistics, programming, and domain knowledge.")
c.drawString(72, 670, "This second PDF is used for batch conversion.")
c.save()

# HTML
html_file = DATA_DIR / "sample_page.html"
html_file.write_text(
    """
    <html>
      <head><title>Docling HTML Example</title></head>
      <body>
        <h1>Docling HTML Example</h1>
        <p>This is a local HTML file converted using Docling.</p>
        <ul>
          <li>PDF conversion</li>
          <li>HTML conversion</li>
          <li>Markdown export</li>
        </ul>
      </body>
    </html>
    """,
    encoding="utf-8"
)

# Markdown
md_file = DATA_DIR / "sample_notes.md"
md_file.write_text(
    """
# Docling Markdown Example

This markdown file is used in the multi-format conversion task.

## Topics
- Document conversion
- Markdown export
- JSON export
    """,
    encoding="utf-8"
)

# CSV
csv_file = DATA_DIR / "sample_table.csv"
pd.DataFrame({
    "name": ["Ali", "Sara", "Omar"],
    "score": [90, 95, 88]
}).to_csv(csv_file, index=False)

# DOCX
docx_file = DATA_DIR / "sample_doc.docx"
doc = Document()
doc.add_heading("Docling DOCX Example", level=1)
doc.add_paragraph("This DOCX file is used to test multi-format conversion.")
doc.add_paragraph("Docling can convert different document types into structured output.")
doc.save(docx_file)

print("Created files:")
for file in DATA_DIR.iterdir():
    print("-", file)
```

```python
converter = DocumentConverter()

source = pdf1

start = time.time()
simple_result = converter.convert(source)
end = time.time()

print(f"Converted in {end - start:.2f} seconds")
print("Status:", simple_result.status)
```

```python
simple_markdown = simple_result.document.export_to_markdown()

print(simple_markdown[:1500])
```

```python
simple_md_path = OUTPUT_DIR / "task1_simple_conversion.md"
simple_json_path = OUTPUT_DIR / "task1_simple_conversion.json"

simple_md_path.write_text(simple_markdown, encoding="utf-8")
simple_json_path.write_text(
    json.dumps(simple_result.document.export_to_dict(), indent=2, ensure_ascii=False),
    encoding="utf-8"
)

print("Saved:")
print(simple_md_path)
print(simple_json_path)
```

```python
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions
from docling.document_converter import PdfFormatOption

pipeline_options = PdfPipelineOptions()

# لأن ملفاتنا PDF نصية وليست scan، نخلي OCR = False عشان يكون أسرع
pipeline_options.do_ocr = False

# تفعيل محاولة فهم الجداول
pipeline_options.do_table_structure = True

custom_converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(
            pipeline_options=pipeline_options
        )
    }
)

custom_result = custom_converter.convert(pdf1)

print("Status:", custom_result.status)
```

```python
custom_md = custom_result.document.export_to_markdown()

custom_md_path = OUTPUT_DIR / "task2_custom_conversion.md"
custom_json_path = OUTPUT_DIR / "task2_custom_conversion.json"

custom_md_path.write_text(custom_md, encoding="utf-8")
custom_json_path.write_text(
    json.dumps(custom_result.document.export_to_dict(), indent=2, ensure_ascii=False),
    encoding="utf-8"
)

print(custom_md[:1500])
print("\nSaved:")
print(custom_md_path)
print(custom_json_path)
```

```python
from docling.datamodel.base_models import ConversionStatus

batch_pdf_files = [pdf1, pdf2]

batch_converter = DocumentConverter()

start = time.time()
batch_results = list(
    batch_converter.convert_all(
        batch_pdf_files,
        raises_on_error=False
    )
)
end = time.time()

print(f"Batch converted in {end - start:.2f} seconds")

for res in batch_results:
    print(res.input.file.name, "=>", res.status)
```

```python
batch_output_dir = OUTPUT_DIR / "task3_batch_conversion"
batch_output_dir.mkdir(exist_ok=True)

success_count = 0
failure_count = 0
partial_success_count = 0

for res in batch_results:
    file_stem = res.input.file.stem

    if res.status == ConversionStatus.SUCCESS:
        success_count += 1

        # Markdown
        (batch_output_dir / f"{file_stem}.md").write_text(
            res.document.export_to_markdown(),
            encoding="utf-8"
        )

        # Plain text
        (batch_output_dir / f"{file_stem}.txt").write_text(
            res.document.export_to_markdown(strict_text=True),
            encoding="utf-8"
        )

        # JSON
        (batch_output_dir / f"{file_stem}.json").write_text(
            json.dumps(res.document.export_to_dict(), indent=2, ensure_ascii=False),
            encoding="utf-8"
        )

        # YAML
        (batch_output_dir / f"{file_stem}.yaml").write_text(
            yaml.safe_dump(res.document.export_to_dict(), allow_unicode=True),
            encoding="utf-8"
        )

    elif res.status == ConversionStatus.PARTIAL_SUCCESS:
        partial_success_count += 1
    else:
        failure_count += 1

print("Success:", success_count)
print("Partial success:", partial_success_count)
print("Failure:", failure_count)

print("\nSaved files:")
for file in batch_output_dir.iterdir():
    print("-", file)
```

```python
from docling.document_converter import WordFormatOption
from docling.pipeline.simple_pipeline import SimplePipeline
from docling.pipeline.standard_pdf_pipeline import StandardPdfPipeline

multi_converter = DocumentConverter(
    allowed_formats=[
        InputFormat.PDF,
        InputFormat.DOCX,
        InputFormat.HTML,
        InputFormat.CSV,
        InputFormat.MD,
    ],
    format_options={
        InputFormat.PDF: PdfFormatOption(
            pipeline_cls=StandardPdfPipeline
        ),
        InputFormat.DOCX: WordFormatOption(
            pipeline_cls=SimplePipeline
        ),
    }
)
```

```python
multi_input_files = [
    pdf1,
    html_file,
    md_file,
    csv_file,
    docx_file,
]

multi_results = list(multi_converter.convert_all(multi_input_files, raises_on_error=False))

for res in multi_results:
    print(res.input.file.name, "=>", res.status)
```

```python
multi_output_dir = OUTPUT_DIR / "task4_multi_format"
multi_output_dir.mkdir(exist_ok=True)

for res in multi_results:
    file_stem = res.input.file.stem

    try:
        markdown_output = res.document.export_to_markdown()
        json_output = res.document.export_to_dict()

        (multi_output_dir / f"{file_stem}.md").write_text(
            markdown_output,
            encoding="utf-8"
        )

        (multi_output_dir / f"{file_stem}.json").write_text(
            json.dumps(json_output, indent=2, ensure_ascii=False),
            encoding="utf-8"
        )

        (multi_output_dir / f"{file_stem}.yaml").write_text(
            yaml.safe_dump(json_output, allow_unicode=True),
            encoding="utf-8"
        )

        print("Saved:", file_stem)

    except Exception as e:
        print("Could not save:", file_stem, "| Error:", e)

print("\nOutput files:")
for file in multi_output_dir.iterdir():
    print("-", file)
```

```python
def to_dict_safely(obj):
    if hasattr(obj, "model_dump"):
        return obj.model_dump()
    if hasattr(obj, "dict"):
        return obj.dict()
    return obj.__dict__

confidence_dict = to_dict_safely(confidence)

confidence_dict
```

```python
def get_field(data, name):
    if isinstance(data, dict):
        return data.get(name)
    return getattr(data, name, None)

print("Document-level confidence grades")
print("--------------------------------")

fields_to_check = [
    "mean_grade",
    "low_grade",
    "layout_grade",
    "ocr_grade",
    "parse_grade",
    "table_grade",
]

for field in fields_to_check:
    print(field, "=>", get_field(confidence_dict, field))
```

```python
print("Document-level confidence scores")
print("--------------------------------")

score_fields = [
    "mean_score",
    "low_score",
    "layout_score",
    "ocr_score",
    "parse_score",
    "table_score",
]

for field in score_fields:
    print(field, "=>", get_field(confidence_dict, field))
```

```python
pages_confidence = get_field(confidence_dict, "pages")

def convert_to_dict(obj):
    if hasattr(obj, "model_dump"):
        return obj.model_dump()
    if hasattr(obj, "dict"):
        return obj.dict()
    return obj

def print_scores_and_grades(obj, prefix=""):
    obj = convert_to_dict(obj)

    if isinstance(obj, dict):
        for key, value in obj.items():
            key_text = str(key)

            if "grade" in key_text.lower() or "score" in key_text.lower():
                print(prefix + key_text, "=>", value)

            elif isinstance(value, (dict, list, tuple)) or hasattr(value, "model_dump"):
                print_scores_and_grades(value, prefix + key_text + ".")

    elif isinstance(obj, (list, tuple)):
        for index, item in enumerate(obj, start=1):
            print_scores_and_grades(item, prefix + f"{index}.")

if pages_confidence:
    print("Page-level confidence:")

    pages_confidence = convert_to_dict(pages_confidence)

    if isinstance(pages_confidence, dict):
        page_items = pages_confidence.items()
    elif isinstance(pages_confidence, list):
        page_items = enumerate(pages_confidence, start=1)
    else:
        page_items = []

    for page_number, page_conf in page_items:
        print(f"\nPage {page_number}")
        print_scores_and_grades(page_conf)

else:
    print("No page-level confidence found.")
```
