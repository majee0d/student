### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url, sep='\t')
```

```python
chipo.head()
```

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)
```

### Step 4. How many products cost more than $10.00?

```python
products = chipo[chipo['quantity'] == 1][['item_name', 'choice_description', 'item_price']].drop_duplicates()
products_over_10 = products[products['item_price'] > 10]
print(products_over_10.shape[0])
```

### Step 5. What is the price of each item? 
###### print a data frame with only three columns item_name choice_description and product_price

```python
products = chipo[chipo['quantity'] == 1][['item_name', 'choice_description', 'item_price']].drop_duplicates()
products.sort_values(by='item_price', ascending=True)
```

### Step 6. Sort by the name of the item

```python
products.sort_values(by='item_name', ascending=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive = chipo.sort_values(by='item_price', ascending=False).head(1)
print(most_expensive[['item_name', 'choice_description', 'quantity', 'item_price']])
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowl_orders = chipo[chipo['item_name'] == 'Veggie Salad Bowl'].shape[0]
print(veggie_salad_bowl_orders)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
more_than_one_canned_soda = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)].shape[0]
print(more_than_one_canned_soda)
```
